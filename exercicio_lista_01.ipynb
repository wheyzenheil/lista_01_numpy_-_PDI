### Codigo limpo e Comentado


```python
import numpy as np
import matplotlib.pyplot as plt
array=np.random.random((800,800))#matriz gerada de forma aleatoria de um tamnho de 800x800
matriz = array.astype(np.float32) #transformando os dados da raiz em float 32
print (matriz)#saida da matriz em float 32
print ('tipo de dado: ',matriz.dtype)#saida tipo de dado da matriz
print ('valor minimo:',np.min(matriz))#saida do valor minimo gerado
print ('valor maximo: ',np.max(matriz))#saida do valor maximo gerado
print ('valor medio:',np.mean(matriz))#saida do valor medio gerado
print ('valor variancia: ', np.var(matriz))#saida da variancia dos valores gerados
#pedaco matriz (retirando uma matriz de dentro de outra)
print('matriz ruido: ')#nome atribuido a matriz
ruido=matriz[0:263,0:264]#o ruido vai ser uma matriz da matriz que pega o valor desde a linha 0 ate a linha 263 e a coluna 0 ate linha 264(valor pode ser alterado contanto que a matriz final tenha 263x264 tamanho)
ruido=np.multiply(ruido,25)#os valores da matriz ruido foram multiplicadas por 25 conforme pedido pelo exercicio

print('foto ruido:')
plt.imshow(ruido, cmap='gray', vmin=0, vmax=255) #saida da foto projetada pela matriz ruido, como o colab so permite 1 foto retirar o # do inicio da linha de codigo
plt.show()
#imagem que sera utilizada na juncao das fotos
print('leitura foto da rosa:') #saida texto antes da foto da rosa
img1 = plt.imread("rose_gray.tif") #leitura da foto da rosa
plt.imshow(img1, cmap='gray') #saida da foto da rosa
plt.show()
#print(img1.shape)#saida tamanho da foto em pixels

#soma da imagem com ruido
res = img1 + ruido #resultado vai ser igual a img1(rosa)+ a imagem ruido gerada pela matriz
print('soma das fotos ruido e da rosa')
plt.imshow(res, cmap='gray' )#saida do resultado
plt.show()
#alterando pixels da foto
#img1.setflags(write=1)#tentativa desbloqueio para edicao explicado nas linhas abaixo
ruido[132,263]=0#pixel da posicao declarada tem valor 0 ou seja desligado
#img1[132,263]=0#pixel na posicao declarada tem valor 0 ou seja desligado(nao foi possivel alterar o pixel da imagem por ser bloqueado )
print('O pixel ruidoso [132,264] tem o valor de:0')#saida em texto do que ocorreu na linha acima
#print('O pixel sem ruido [132,264] tem o valor de: 0')#saida em texto do que ocorreu na linha acima
plt.imshow(ruido, cmap='gray')#saida da foto com os pixels alterados
plt.show()




```
